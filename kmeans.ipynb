```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dataset=pd.read_csv('kss.csv')
```

```python
x=dataset.iloc[:,:].values
```

```python
wscc = []
for i in range(1,11):
    
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wscc.append(kmeans.inertia_)
plt.plot(range(1,11), wscc)
plt.title('its elbow model')
plt.xlabel('no of cluster')
plt.ylabel('wscc')
plt.show()
    

    
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42)
y_kmean=kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_kmean == 0,0],x[y_kmean == 0,1],s=100 , c='blue' , label='cluster 1')
plt.scatter(x[y_kmean == 1,0],x[y_kmean == 1,1],s=100 , c='orange' , label='cluster 2')
plt.scatter(x[y_kmean == 2,0],x[y_kmean == 2,1],s=100 , c='red' , label='cluster 3')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200 , c='pink' , label='centroids')

plt.title('no of custumers')
plt.xlabel('annual income $')
plt.ylabel('spending score 1-100')
```
